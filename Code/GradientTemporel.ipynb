```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def temporal_gradient(images):
    n, height, width = images.shape
    gradients = np.zeros((n-1, height, width))
    
    for i in range(1, n):
        gradients[i-1] = images[i] - images[i-1]
    
    return gradients

# Charger la squence d'images
image_sequence = []
for i in range(1, 9):  # Suppose que vous avez des images i0001.png  i0010.png
    image_path = "MiniCooper/i{:04d}.png".format(i)
    image = cv2.imread(image_path, 0) / 255.0
    image_sequence.append(image)

# Convertir en tableau numpy
image_sequence = np.array(image_sequence)

# Calculer les gradients temporels
temporal_gradients = temporal_gradient(image_sequence)
```

```python

# Afficher les deux premires images de la squence et le gradient temporel correspondant
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.imshow(image_sequence[0], cmap='gray')
plt.title('Image 1')

plt.subplot(132)
plt.imshow(image_sequence[1], cmap='gray')
plt.title('Image 2')

plt.subplot(133)
plt.imshow(temporal_gradients[0], cmap='gray')
plt.title('Gradient Temporel (I2 - I1)')

plt.show()
```
